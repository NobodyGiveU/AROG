```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path = '/Users/kavithasubham/Downloads/Ben/Python Plotting 1 - Injury Risk.xlsx'

sheets = pd.read_excel(file_path, sheet_name=None)

plot1_df = sheets['Plot 1 - Age']
plot2_df = sheets['Plot 2 - Height']
plot3_df = sheets['Plot 3 - Weight']
plot4_df = sheets['Plot 4 - Training Intensity']
plot5_df = sheets['Plot 5 - Training Hours']
plot6_df = sheets['Plot 6 - Rest Days']
plot7_df = sheets['IMPORTANT Plot - Gender']

#print(plot1_df.head(100))
```

```python
#Plot1
plt.figure(figsize=(10, 6))

# Plot actual data lines
plt.plot(plot1_df['Age'], plot1_df['Female'], color='blue', label='Female')
plt.plot(plot1_df['Age'], plot1_df['Male'], color='red', label='Male')
plt.plot(plot1_df['Age'], plot1_df['Overall'], color='orange', label='Overall')

# Add regression (trend) lines
sns.regplot(x='Age', y='Female', data=plot1_df, scatter=False, color='blue', ci=None, line_kws={'alpha': 0.4})
sns.regplot(x='Age', y='Male', data=plot1_df, scatter=False, color='red', ci=None, line_kws={'alpha': 0.4})
sns.regplot(x='Age', y='Overall', data=plot1_df, scatter=False, color='orange', ci=None, line_kws={'alpha': 0.4})

# Labels and title
plt.xlabel('Age')
plt.ylabel('Number of ACL Injuries')
plt.title('ACL Injury Rates Vs. Age')
plt.xticks(plot1_df['Age'])
plt.grid(True)
plt.legend()
plt.savefig("Fig_1.pdf", dpi=600, bbox_inches='tight')
plt.show()
```

```python
file_path = '/Users/kavithasubham/Downloads/Ben/Python Plotting 2 - Recovery Time.xlsx'

sheets = pd.read_excel(file_path, sheet_name=None)

plot1_df = sheets['Plot 1 - Age']
plot2_df = sheets['Plot 2 - Height']
plot3_df = sheets['Plot 3 - Weight']
plot4_df = sheets['Plot 4 - Jump Height']
plot5_df = sheets['Plot 5 - Speed']
plot6_df = sheets['Plot 6 - Rehabilitation Efficie']
plot7_df = sheets['Extra Plot - Rehab Program']

#print(plot1_df.head(100))
```

```python
#Plot8
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Line plot
sns.lineplot(x='Age', y='Average Recovery Time', data=plot1_df, label="Recovery Time", color='blue')

# Regression line (trendline)
z = np.polyfit(plot1_df['Age'], plot1_df['Average Recovery Time'], 1)
p = np.poly1d(z)
plt.plot(plot1_df['Age'], p(plot1_df['Age']), color='skyblue', linestyle='-', linewidth=2, alpha=0.7, label="Trendline")

# Labels and title
plt.title("ACL Recovery Time Vs. Age", fontsize=14, fontweight='bold')
plt.xlabel("Age", fontsize=12)
plt.ylabel("Average ACL Recovery Time in Weeks", fontsize=12)
plt.xticks(plot1_df['Age'])
plt.legend()
plt.tight_layout()
plt.savefig("Fig_8.pdf", dpi=600, bbox_inches='tight')
plt.show()
```
